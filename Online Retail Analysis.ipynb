```python
# Loading the data set
import pandas as pd
import openpyxl
```

```python
df = pd.read_excel(r"Online Retail.xlsx")
df
```

```python
df.head()
```

```python
df = df.copy()
```

```python
## Handling missing data
print(df.isnull().sum())
```

```python
## Filling in Description with unknown
df.dropna(subset = ["Description", "CustomerID"], inplace = True)
```

```python
## Removing duplicates
df = df.drop_duplicates()
df
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
## Checking data stypes
print(df.dtypes)
```

```python
## Separating date and time from invoice date column
## Creating a new date column
df["InvoiceDate_Only"] = df["InvoiceDate"].dt.date
```

```python
## Creating a new time column
df["InvoiceTime_only"] = df["InvoiceDate"].dt.time
```

```python
## Checking to see if it worked
df.head()
```

```python
## Compuing new features
### Adding a new total price column
df.loc[:,"Total_Price"] = df["Quantity"] * df["UnitPrice"]
```

```python
## Ensuring invoicedae column is in the right format
df.loc[:,"InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
## Extract year, month, day and hour
df.loc[:,"Year"] = df["InvoiceDate"].dt.year
df.loc[:,"Month"] = df["InvoiceDate"].dt.month
df.loc[:,"Day"] = df["InvoiceDate"].dt.day
df.loc[:,"Hour"] = df["InvoiceDate"].dt.hour
```

```python
df.head()
```

```python
# Additional date features
df.loc[:,"WeekNumber"] = df["InvoiceDate"].dt.isocalendar().week
df.loc[:,"DayOfWeek"] = df["InvoiceDate"].dt.day_name()  # e.g., Monday, Tuesday
df.loc[:,"IsWeekend"] = df["DayOfWeek"].isin(["Saturday", "Sunday"])  # Boolean flag
```

```python
print(df[["InvoiceDate", "Year", "Month", "Day", "Hour", "WeekNumber", "DayOfWeek", "IsWeekend"]].head())
```
